```python
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
#Extract Data From the PDF File
def load_pdf_file(data):
    loader= DirectoryLoader(data,
                            glob="*.pdf",
                            loader_cls=PyPDFLoader)

    documents=loader.load()

    return documents
```

```python
!pip install pypdf
```

```python
extracted_data=load_pdf_file(data='content/Data')
```

```python
extracted_data
```

```python
#Split the Data into Text Chunks
def text_split(extracted_data):
    text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
    text_chunks=text_splitter.split_documents(extracted_data)
    return text_chunks
```

```python
text_chunks=text_split(extracted_data)
print("Length of Text Chunks", len(text_chunks))
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer
import torch
import warnings
import transformers
warnings.filterwarnings("ignore")
# Load Llama2 Model

model = "daryl149/llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model)
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True,
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline=pipeline,model_kwargs={'temperature':0,'max_new_tokens':1500})

prompt="What would be a good name for a company that integrates web dev projects with AI"

print(llm(prompt))
```

```python
from langchain.chains import RetrievalQA
query="What is Self Attention Mechanism?"
retrieved_docs = retriever.invoke(query)
```

```python
retrieved_docs
```

```python
chain = RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriever)
```

```python
query="What is Self Attention Mechanism?"
response = chain.run(query)
print(response)
```
