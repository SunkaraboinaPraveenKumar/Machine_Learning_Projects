Text generation is a NLP task that involves creating coherent and contextually relevant textual content. It can be creative and utilitarian, finding applications in various domains

Highest Probable next text can be generated by Greedy Search, Beam Search, Top-K Sampling, Top-p(nucleus) Sampling

Challenges in generating coherent text :
Repetition: can repeat itself generating same text over and over again
Limited Vocab:model using same words or phrases over and over again
Lack of Context: grammatically correct but lack context and meaning

```python
import torch
from transformers import AutoTokenizer,AutoModelForCausalLM
```

```python
device="cuda" if torch.cuda.is_available() else "cpu"
```

```python
device
```

```python
model_name="openai-community/gpt2-large"
tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
model=AutoModelForCausalLM.from_pretrained(model_name).to(device)
```

```python
input_text="I am really Impressed"
max_length=128
input_ids=tokenizer.encode(input_text,return_tensors="pt")
```

```python
input_ids
```

Beam Search

```python
output=model.generate(input_ids.to(device),max_length=max_length,num_beams=5,no_repeat_ngram_size=2,do_sample=False,early_stopping=True)
```

```python
print(tokenizer.decode(output[0]))
```

Nuclear Search

```python
input_text="Welcome to JNTU College of Engineering"
max_length=128
input_ids=tokenizer.encode(input_text,return_tensors="pt")
```

```python
output=model.generate(input_ids.to(device),max_length=max_length,do_sample=False,top_p=0.92,no_repeat_ngram_size=2)
tokenizer.decode(output[0])
```
