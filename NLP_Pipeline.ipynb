```python
from tika import parser

def extract_text_from_pdf(pdf_path):
    try:
      parsed_pdf=parser.from_file(pdf_path)
      text = parsed_pdf.get("content","")
      if text:
        return text.strip()
      else:
        return "NO text could be extracted from the file"
    except Exception as e:
      return f"An error occurred: {str(e)}"

text = extract_text_from_pdf('/content/🔺 PYTHON 🔺  CHEATSHEETS  ☂️COLLECTION.pdf')
print(text)
```

```python
import pandas as pd
from snorkel.labeling import labeling_function, PandasLFApplier
from snorkel.labeling.model import LabelModel
```

```python
ABSTAIN=-1
POSITIVE=1
NEGATIVE=0
```

```python
@labeling_function()
def lf_contains_good(x):
  return POSITIVE if "good" or "best" or "great" in x.text.lower() else ABSTAIN

@labeling_function()
def lf_contains_bad(x):
  return NEGATIVE if "bad" or "worst" or "terrible" in x.text.lower() else ABSTAIN

@labeling_function()
def lf_contains_average(x):
  return NEGATIVE if "average" in x.text.lower() else ABSTAIN
```

```python
data = pd.DataFrame({
    "text":[
        "This movie is good",
        "This movie is bad",
        "An average flim, nothing special",
        "I love NLP, LLMs and CV",
        "I hate NLP, LLMs and CV",
        "NTR and KTR are best future of AP and TS",
        "Hyderabad is a good city",
        "Do you know this?"
    ]
})
```

```python
lfs=[lf_contains_good, lf_contains_bad, lf_contains_average]
applier=PandasLFApplier(lfs=lfs)
L_train=applier.apply(data)

label_model=LabelModel(cardinality=2, verbose=True)
label_model.fit(L_train, n_epochs=500,log_freq=100,seed=123)
data["label"]=label_model.predict(L_train)

data.to_csv("labelled_data.csv",index=False)
print("Weakly supervised labels:")
print(data)
```
