```python
import numpy as np
```

Perceptron Class Sigmoid

```python
class SigmoidPerceptron():

  def __init__(self,input_size):
    self.weights=np.random.randn(input_size)
    self.bias=np.random.randn(1)
  def sigmoid(self,z):
    return 1/(1+np.exp(-z))
  def predict(self,inputs):
    weighted_sum=np.dot(inputs,self.weights)+self.bias
    return self.sigmoid(weighted_sum)
  def fit(self,inputs,targets,learning_rate,num_of_epochs):
    #optimization
    num_examples=inputs.shape[0]
    for epoch in range(num_of_epochs):
      #stochastic gradient descent for each individual data points
      for i in range(num_examples):
        input_vector=inputs[i]

        target=targets[i]

        prediction=self.predict(input_vector)

        error=target-prediction

        #update weights
        gradient_weights=error*prediction*(1-prediction)*input_vector
        self.weights+=learning_rate*gradient_weights
        #update bias
        gradient_bias=error*prediction*(1-prediction)
        self.bias+=learning_rate*gradient_bias

  #evaluate model
  def evaluate(self,inputs,targets):
    correct=0
    for input_vector,target in zip(inputs,targets):
      prediction=self.predict(input_vector)

      if prediction>=0.5:
        predicted_class=1
      else:
        predicted_class=0
      if predicted_class==target:
        correct+=1
    accuracy=correct/len(inputs) #no.of correct prediction/total no.of data points
    return accuracy
```

Importing the dependencies

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

Imbalance dataset

```python
class_0_df=df[df['Outcome']==0]
class_1_df=df[df['Outcome']==1]

class_0_df=class_0_df.sample(268)
data=pd.concat([class_0_df,class_1_df])
```

Train Test Split

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
model=SigmoidPerceptron(input_size=X_train_scaled.shape[1])
```

```python
#train the perceptron model
model.fit(inputs=X_train_scaled,targets=Y_train,learning_rate=0.01,num_of_epochs=1000)
```

Model Evaluation

```python
train_accuracy=model.evaluate(X_test_scaled,Y_test)
test_accuracy=model.evaluate(X_train_scaled,Y_train)
train_accuracy=round(train_accuracy*100,2)
test_accuracy=round(test_accuracy*100,2)
print('Training Accuracy:',train_accuracy,'%')
print('Test Accuracy:',test_accuracy,'%')
```
