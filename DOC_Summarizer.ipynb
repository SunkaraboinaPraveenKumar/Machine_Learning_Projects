```python
from llama_index.core import SimpleDirectoryReader
from llama_index.core import SummaryIndex
from llama_index.llms.groq import Groq
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core import Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
```

```python
llm=Groq(model="llama-3.1-8b-instant")
Settings.llm = llm
Settings.embed_model=HuggingFaceEmbedding()
```

```python
# load the document
document = SimpleDirectoryReader(input_files=["/content/Open_Source_AI-Enhanced_3D-Printed_Insulin_Pump.pdf"]).load_data()
```

```python
print(type(document))
```

```python
len(document)
```

```python
document[0]
```

```python
splitter = SentenceSplitter(chunk_size=2048)
nodes = splitter.get_nodes_from_documents(document)
```

```python
summary_index = SummaryIndex(nodes)
```

```python
summary_query_engine = summary_index.as_query_engine(
    response_mode='tree_summarize',
    use_async=True
)
```

```python
response=summary_query_engine.query("Summarize the given document")
```

```python
print(response)
```

```python
# storing the document summary
summary_index.storage_context.persist('doc_summary_index')
```

```python
response=summary_query_engine.query("Summarize the given document in detail")
```
