````python
from typing import Optional, Dict
from sentence_transformers import SentenceTransformer, util
import spacy
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from transformers import pipeline
import language_tool_python
import textstat
from dataclasses import dataclass

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Load NLP models
spacy_model = spacy.load("en_core_web_sm")
semantic_model = SentenceTransformer('all-MiniLM-L6-v2')
grammar_tool = language_tool_python.LanguageTool('en-US')
````

````python
from dataclasses import dataclass
from typing import Optional

@dataclass
class PipelineState:
    teacher_text: Optional[str] = None
    student_text: Optional[str] = None
    teacher_text_clean: Optional[str] = None
    student_text_clean: Optional[str] = None
    tokenized_words: Optional[list] = None
    grammar_score: Optional[float] = None
    syntactic_score: Optional[float] = None
    semantic_score: Optional[float] = None
    final_score: Optional[float] = None
    explanation: Optional[str] = None
    # New fields for teacher text analysis
    teacher_definition: Optional[str] = None
    teacher_description: Optional[str] = None
    teacher_conclusion: Optional[str] = None
    # New fields for student text analysis
    student_definition: Optional[str] = None
    student_description: Optional[str] = None
    student_conclusion: Optional[str] = None
    teacher_keywords: Optional[str] = None
    student_keywords: Optional[str] = None
    llm_feedback: Optional[str] = None
````

````python
import re

def text_cleaning_node(state: PipelineState) -> PipelineState:
    def clean_text(text: str) -> str:
        if not text:
            return ""
        # Replace all whitespace (including newlines, tabs) with a single space
        cleaned = re.sub(r'\s+', ' ', text)
        return cleaned.strip()

    state.teacher_text_clean = clean_text(state.teacher_text)
    state.student_text_clean = clean_text(state.student_text)
    return state
````

````python
def keyword_matching_node(state: PipelineState) -> PipelineState:
    """
    Check which teacher keywords are present in the student text.
    Assumes state.teacher_keywords is a comma-separated string of keywords.
    """
    # If teacher_keywords is not provided or student text is empty, update student_keywords to an empty string.
    if not state.teacher_keywords or not state.student_text_clean:
        state.student_keywords = ""
        return state

    # Split teacher keywords by comma, strip spaces, and lowercase them.
    teacher_kw_list = [kw.strip().lower() for kw in state.teacher_keywords.split(",") if kw.strip()]

    # Lowercase the student text for case-insensitive matching.
    student_text_lower = state.student_text_clean.lower()

    # Check each teacher keyword to see if it appears in the student text.
    matched_keywords = [kw for kw in teacher_kw_list if kw in student_text_lower]

    # Update student_keywords with the matched keywords as a comma-separated string.
    state.student_keywords = ", ".join(matched_keywords)

    print(f"Matched Teacher Keywords in Student Text: {state.student_keywords}")
    return state
````

````python
from langchain_groq import ChatGroq
import re

llm = ChatGroq(model="llama-3.3-70b-versatile")  # Adjust model as needed

def extract_sections_from_text(text: str) -> dict:
    """
    Uses an LLM to extract definition, description, and conclusion from the given text.
    """
    if not text:
        return {}

    prompt = f"""
    Analyze the following text and extract any of these sections if they are explicitly present:
    - **Definition**: A clear statement defining the key concept.
    - **Description**: Detailed explanation of the concept, including examples or elaboration.
    - **Conclusion**: A final summarizing statement about the concept.

    **Text:**
    {text}

    **Instructions:**
    - Only include a section in your output if it is explicitly present in the text.
    - Do not include any placeholder or empty sections. For example, if the definition or description is missing, do not output anything for them.
    - If none of the sections are present, return an empty response.

    **Output Format:**
    For each section present, output in this exact format:
    ```
    SectionName: <Extracted Content>
    ```
    For example, if only the conclusion is present, your output should be:
    ```
    Conclusion: <Extracted Conclusion>
    ```
    """
    response = llm.invoke(prompt).content

    # Extract relevant parts using regex
    definition = re.search(r"Definition:\s*(.+)", response)
    description = re.search(r"Description:\s*(.+)", response)
    conclusion = re.search(r"Conclusion:\s*(.+)", response)

    extracted_data = {}
    if definition:
        extracted_data["definition"] = definition.group(1)
    if description:
        extracted_data["description"] = description.group(1)
    if conclusion:
        extracted_data["conclusion"] = conclusion.group(1)

    return extracted_data



def extract_sections_node(state: PipelineState) -> PipelineState:
    """
    Uses an LLM to extract definition, description, and conclusion for both teacher and student texts.
    """
    teacher_sections = extract_sections_from_text(state.teacher_text_clean or state.teacher_text)
    student_sections = extract_sections_from_text(state.student_text_clean or state.student_text)

    # Update state with extracted values for both teacher and student
    state.teacher_definition = teacher_sections.get("definition",None)
    state.teacher_description = teacher_sections.get("description",None)
    state.teacher_conclusion = teacher_sections.get("conclusion",None)

    state.student_definition = student_sections.get("definition",None)
    state.student_description = student_sections.get("description",None)
    state.student_conclusion = student_sections.get("conclusion",None)

    return state
````

````python
from nltk.tokenize import sent_tokenize
import textstat

def count_sentences(text: str) -> int:
    """Returns the number of sentences in a given text."""
    return len(sent_tokenize(text)) if text else 0

def normalize_readability(score: float, min_score: float = 3.0, max_score: float = 20.0) -> float:
    """Normalizes readability scores to a 0-1 range where 1 is best."""
    if score < min_score:
        return 1.0
    return max(0.0, min(1.0, 1 - (score - min_score)/(max_score - min_score)))

def extended_syntactic_analysis_node(state: PipelineState) -> PipelineState:
    # Ensure text is cleaned
    teacher_text = state.teacher_text_clean or state.teacher_text or ""
    student_text = state.student_text_clean or state.student_text or ""

    # --- Readability Analysis ---
    try:
        smog_teacher = textstat.smog_index(teacher_text)
        ari_teacher = textstat.automated_readability_index(teacher_text)
    except Exception as e:
        print(f"Error calculating teacher readability: {e}")
        smog_teacher, ari_teacher = 0, 0

    try:
        smog_student = textstat.smog_index(student_text)
        ari_student = textstat.automated_readability_index(student_text)
    except Exception as e:
        print(f"Error calculating student readability: {e}")
        smog_student, ari_student = 0, 0

    # Normalize readability scores (higher is better)
    norm_smog_teacher = normalize_readability(smog_teacher)
    norm_ari_teacher = normalize_readability(ari_teacher)
    norm_smog_student = normalize_readability(smog_student)
    norm_ari_student = normalize_readability(ari_student)

    readability_teacher = (norm_smog_teacher + norm_ari_teacher) / 2
    readability_student = (norm_smog_student + norm_ari_student) / 2
    readability_score = (readability_teacher + readability_student) / 2

    # --- Section Extraction ---
    state = extract_sections_node(state)

    # --- Sentence Count Similarity ---
    teacher_counts = {
        "definition": count_sentences(state.teacher_definition),
        "description": count_sentences(state.teacher_description),
        "conclusion": count_sentences(state.teacher_conclusion)
    }

    student_counts = {
        "definition": count_sentences(state.student_definition),
        "description": count_sentences(state.student_description),
        "conclusion": count_sentences(state.student_conclusion)
    }

    # Calculate sentence match ratio (0-1 scale)
    sentence_match_ratio = sum(
        min(student_counts[sec], teacher_counts[sec]) / max(1, teacher_counts[sec])
        for sec in ["definition", "description", "conclusion"]
    ) / 3  # Average across sections

    # --- Section Presence Evaluation ---
    section_presence_score = sum(
        1 if section else 0
        for section in [state.student_definition, state.student_description, state.student_conclusion]
    ) / 3

    # --- Final Score Calculation ---
    state.syntactic_score = (
        (0.4 * readability_score) +          # 40% readability
        (0.4 * sentence_match_ratio) +       # 40% sentence structure match
        (0.2 * section_presence_score)       # 20% section presence
    )

    print(f"\n\nTeacher Sections: Definition: {state.teacher_definition},\n\n Description: {state.teacher_description}, \n\nConclusion: {state.teacher_conclusion}")
    print(f"\n\nStudent Sections: Definition: {state.student_definition},\n\n Description: {state.student_description},\n\n Conclusion: {state.student_conclusion}")
    print(f"\n\nSentence Match Ratio: {sentence_match_ratio:.2f}, \n\nSection Presence Score: {section_presence_score:.2f}")
    print(f"\nFinal Adjusted Syntactic Score: {state.syntactic_score:.2f}")

    return state
````

````python
import math
def grammar_analysis(state: PipelineState) -> PipelineState:
    """Analyzes grammatical correctness using LanguageTool."""
    matches = grammar_tool.check(state.student_text_clean)
    num_errors = len(matches)
    total_words = max(len(state.student_text_clean.split()), 1)

    # Normalize error count to a score between 0 and 1
    normalized_errors = num_errors / total_words
    state.grammar_score = max(0, min(1, 1 - math.log1p(normalized_errors)))  # log1p(x) = log(1 + x)

    return state
````

````python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sentence_transformers import util

# Ensure stopwords are downloaded
stop_words = set(stopwords.words('english'))

def remove_stopwords(text: str) -> str:
    """
    Tokenizes the text, removes stopwords, and reconstructs the text.
    """
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    return " ".join(filtered_tokens)

def semantic_analysis(state: PipelineState) -> PipelineState:
    """
    Computes semantic similarity after removing stopwords from teacher and student text.
    """
    # Remove stopwords from both texts
    teacher_cleaned = remove_stopwords(state.teacher_text_clean)
    student_cleaned = remove_stopwords(state.student_text_clean)

    # Generate embeddings
    teacher_embedding = semantic_model.encode(teacher_cleaned, convert_to_tensor=True)
    student_embedding = semantic_model.encode(student_cleaned, convert_to_tensor=True)

    # Compute cosine similarity
    state.semantic_score = util.pytorch_cos_sim(teacher_embedding, student_embedding).item()

    return state
````

````python
def final_evaluation(state: PipelineState) -> PipelineState:
    """
    Computes the final score based on weighted factors:
    - Semantic Similarity: 50%
    - Syntactic Accuracy: 30%
    - Grammar Accuracy: 20%
    """
    state.final_score = (
        (0.5 * state.semantic_score) +
        (0.3 * state.syntactic_score) +
        (0.2 * state.grammar_score)
    )
    return state
````

````python
def llm_feedback_node(state: PipelineState) -> PipelineState:
    """
    Uses an LLM to provide feedback based on weighted scores and section completeness,
    comparing the teacher's and student's answers.
    """
    # Determine which sections are missing in the student's answer.
    missing_sections = []
    if state.student_definition is None:
        missing_sections.append("Definition")
    if state.student_description is None:
        missing_sections.append("Description")
    if state.student_conclusion is None:
        missing_sections.append("Conclusion")

    missing_text = ", ".join(missing_sections) if missing_sections else "None"

    prompt = f"""
    Given the following analysis of a student's answer:

    - **Semantic Similarity Score** (50% weight): {state.semantic_score}
    - **Syntactic Score** (30% weight): {state.syntactic_score}
    - **Grammar Score** (20% weight): {state.grammar_score}

    Additionally, note that the student's answer is missing the following section(s): {missing_text}.

    The teacher's analysis of the text includes:
    - Definition: {state.teacher_definition}
    - Description: {state.teacher_description}
    - Conclusion: {state.teacher_conclusion}

    The student's answer includes:
    - Definition: {state.student_definition}
    - Description: {state.student_description}
    - Conclusion: {state.student_conclusion}

    If none of the sections are present, return a score of 0.

    Please generate a final score (0-100) and provide **detailed feedback** on how the student can improve their answer. In your feedback, include suggestions on how to address the missing sections (if any) and improve overall quality.
    """

    response = llm.invoke(prompt)
    state.llm_feedback = response.content
    return state
````

````python
# Define the graph
import langgraph
from langgraph.graph import StateGraph
graph = StateGraph(PipelineState)

# Add nodes (processing steps)
graph.add_node("preprocess", text_cleaning_node)
graph.add_node("keyword_matching", keyword_matching_node)
graph.add_node("extract_sections", extract_sections_node)
graph.add_node("syntactic", extended_syntactic_analysis_node)
graph.add_node("grammar", grammar_analysis)
graph.add_node("semantic", semantic_analysis)
graph.add_node("final_evaluation", final_evaluation)
graph.add_node("llm_feedback_final", llm_feedback_node)

# Define execution flow
graph.add_edge("preprocess", "keyword_matching")
graph.add_edge("keyword_matching", "extract_sections")     # Section Extraction → Semantic Analysis
graph.add_edge("extract_sections", "syntactic")     # Semantic → Keyword Matching
graph.add_edge("syntactic", "grammar")            # Semantic → Syntactic Analysis
graph.add_edge("grammar", "semantic")             # Syntactic → Grammar Analysis
graph.add_edge("semantic", "final_evaluation")      # Grammar → Final Score Calculation
graph.add_edge("final_evaluation", "llm_feedback_final")  # Final Score → LLM Feedback

# Set entry point
graph.set_entry_point("preprocess")

# Compile the graph
app = graph.compile()
````
