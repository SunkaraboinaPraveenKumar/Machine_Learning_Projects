```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic=pd.read_csv(url)
```

```python
titanic.head()
```

```python
print(titanic.isnull())
```

```python
print(titanic.isnull().sum())
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(),inplace=True)
```

```python
titanic.info()
```

```python
titanic.drop('Cabin',inplace=True,axis=1)
```

```python
titanic.info()
```

```python
titanic['Survived'].mean()*100
```

```python
titanic.groupby('Sex')['Survived'].mean()*100
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
titanic.groupby('Survived')['Age'].mean()
```

```python
titanic[titanic['Age']<16].groupby('Survived')['Age'].mean()
```

```python
titanic[titanic['Age']<16]['Survived'].mean()*100
```

```python
titanic[titanic['Age']>16]['Survived'].mean()*100
```

```python
titanic[titanic['Survived']==1].sort_values(by='Age',ascending=True)
```

```python
bins = [0,16,60,float('inf')]
labels = ['Children','Adult','Senior']
titanic['AgeGroup'] = pd.cut(titanic['Age'],bins=bins,labels=labels,right=False)
```

```python
titanic.head()
```

```python
titanic.groupby('AgeGroup')['Survived'].mean()*100
```
