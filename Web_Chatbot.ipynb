```python
from langchain.document_loaders import UnstructuredURLLoader

# Define your URLs
URLs = [
    "https://www.databricks.com/blog/mpt-7b",
    "https://stability.ai/blog/stability-ai-launches-the-first-of-its-stablelm-suite-of-language-models",
    "https://jntuh.ac.in/"
]

# Initialize the loader
loaders = UnstructuredURLLoader(urls=URLs)

# Load data from URLs
data = loaders.load()

print(data)
```

```python
len(data)
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator='\n',chunk_size=1000,chunk_overlap=200)
```

```python
text_chunks=text_splitter.split_documents(data)
```

```python
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load Llama2 Model

model = "daryl149/llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
!pip install safetensors
```

```python
import transformers
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True,
    max_length=1200,
    max_new_tokens=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline=pipeline,model_kwargs={'temperature':0,'max_new_tokens':1500})

prompt="What would be a good name for a company that integrates web dev projects with AI"

print(llm(prompt))
```

```python
from langchain.chains import RetrievalQA
query="What is JNTU?"
docs=vectordb.similarity_search(query,k=3)
docs
```

```python
chain = RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=vectordb.as_retriever())
```

```python
query="What is JNTU?"
response = chain.run(query)
print(response)
```
