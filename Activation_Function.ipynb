```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

x, y = make_blobs(n_samples=1000,centers=2, random_state=42, cluster_std=1.5)

scaler = StandardScaler()
x=scaler.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

model = Sequential([
    Dense(10, activation='linear'),
    Dense(10, activation='linear'),
    Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])

history = model.fit(x_train,y_train,epochs=50,batch_size=32,verbose=0)

y_pred=(model.predict(x_test)>0.5).astype('int32')

accuracy=accuracy_score(y_test,y_pred)
print(f"Test Accuracy: {accuracy*100:.2f}%")

def plot_decision_boundary(model, X, y):
  x_min, x_max=X[:,0].min()-1, X[:,0].max()+1
  y_min, y_max=X[:,1].min()-1, X[:,1].max()+1
  xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
  np.arange(y_min, y_max, 0.1))
  Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
  Z = (Z > 0.5).astype(int)
  Z = Z.reshape(xx.shape)
  plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)
  plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
  plt.xlabel('X1')
  plt.ylabel('X2')
  plt.title('Decision Boundary')
  plt.show()

plot_decision_boundary(model, x_test, y_test)
```

Sigmoid/ Logistic Activation

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
plt.style.use('ggplot')

class Sigmoid(nn.Module):
  def __init__(self):
    super(Sigmoid, self).__init__()

  def forward(self, x):
    return 1 / (1+torch.exp(-x))

  def backward(self, x):
    return self.forward(x) * (1-self.forward(x))
```

```python
sigmoid_layer = Sigmoid()
x = torch.linspace(-10,10,100)
```

```python
sigmoid_layer
```

```python
y=sigmoid_layer(x)
```

```python
y
```

```python
plt.plot(x.numpy(),y.numpy())
plt.xlabel('X')
plt.ylabel('Sigmoid of X')
plt.title('Sigmoid Activation Function')
plt.grid(True)
plt.show()
```

Tanh

```python
class Tanh(nn.Module):
  def __init__(self):
    super(Tanh, self).__init__()

  def forward(self, x):
    exp_x = torch.exp(x)
    exp_neg_x = torch.exp(-x)
    return (exp_x - exp_neg_x)/(exp_x+exp_neg_x)
```

```python
sigmoid_layer = Tanh()
x = torch.linspace(-10,10,100)
y=sigmoid_layer(x)
```

```python
y
```

```python
plt.plot(x.numpy(),y.numpy())
plt.xlabel('X')
plt.ylabel('Tanh of X')
plt.title('Tanh Activation Function')
plt.grid(True)
plt.show()
```

ReLU Function

```python
class Relu(nn.Module):
  def __init__(self):
    super(Relu, self).__init__()

  def forward(self, x):
    return torch.maximum(torch.tensor(0.0), x)

relu_layer = Relu()
x = torch.linspace(-10,10,100)
y=relu_layer(x)
```

```python
y
```

```python
plt.plot(x.numpy(),y.numpy())
plt.xlabel('X')
plt.ylabel('ReLU of X')
plt.title('ReLU Activation Function')
plt.grid(True)
plt.show()
```

Leaky Relu

```python
class LeakyRelu(nn.Module):
  def __init__(self, alpha=0.1):
    self.alpha=alpha
    super(LeakyRelu, self).__init__()

  def forward(self, x):
    return torch.where(x>=0, x, self.alpha*x)

relu_layer = LeakyRelu()
x = torch.linspace(-10,10,100)
y=relu_layer(x)
```

```python
y
```

```python
plt.plot(x.numpy(),y.numpy())
plt.xlabel('X')
plt.ylabel('LeakyReLU of X')
plt.title('LeakyReLU Activation Function')
plt.grid(True)
plt.show()
```

Parametric ReLU

```python
class ParametricRelu(nn.Module):
  def __init__(self, alpha=0.1):
    super(ParametricRelu, self).__init__()
    self.alpha=nn.Parameter(torch.tensor(alpha, dtype=torch.float32))

  def forward(self, x):
    return torch.where(x>=0, x, self.alpha*x)

parametric_relu = ParametricRelu()
x = torch.linspace(-10,10,100)
y=parametric_relu(x)
```

```python
y
```

```python
plt.plot(x.detach().numpy(),y.detach().numpy())
plt.xlabel('X')
plt.ylabel('ParametricReLU of X')
plt.title('ParametricReLU Activation Function')
plt.grid(True)
plt.show()
```

Exponential Linear unit (ELUs)

```python
class ExponentialLinearUnit(nn.Module):
  def __init__(self, alpha=1.0):
    super(ExponentialLinearUnit, self).__init__()
    self.alpha=alpha

  def forward(self, x):
    return torch.where(x>=0, x, self.alpha*(torch.exp(x)-1))

expo_layer = ExponentialLinearUnit()
x = torch.linspace(-5,5,200)
y=expo_layer(x)
```

```python
y
```

```python
plt.plot(x.numpy(),y.numpy())
plt.xlabel('X')
plt.ylabel('ELU of X')
plt.title('ELU Activation Function')
plt.grid(True)
plt.show()
```

SELU

```python
class SELU(nn.Module):
  def __init__(self, alpha=1.6733, lambda_=1.0507):
    super(SELU, self).__init__()
    self.lambda_=lambda_
    self.alpha=alpha

  def forward(self, x):
    return torch.where(x>=0, self.lambda_*x, self.lambda_*self.alpha*(torch.exp(x)-1))

selu_layer = SELU()
x = torch.linspace(-5,5,200)
y=selu_layer(x)
```

```python
y
```

```python
plt.plot(x.detach().numpy(),y.detach().numpy())
plt.xlabel('X')
plt.ylabel('SELU of X')
plt.title('SELU Activation Function')
plt.grid(True)
plt.show()
```

Softmax - Combination of multiple Sigmoid

```python
class Softmax(nn.Module):
  def __init__(self):
    super(Softmax, self).__init__()

  def forward(self, x):
    exp_x=torch.exp(x)
    return exp_x/torch.sum(exp_x, dim=0)

softmax_layer = Softmax()
x = torch.linspace(-5,5,100)
y=softmax_layer(x)
```

```python
y
```

```python
plt.plot(x.numpy(),y.numpy())
plt.xlabel('X')
plt.ylabel('Softmax of X')
plt.title('Softmax Activation Function')
plt.grid(True)
plt.show()
```

Swish Activation f(x)=x*sigmoid(x)

```python
class Swish(nn.Module):
  def __init__(self):
    super(Swish, self).__init__()

  def forward(self, x):
    sigmoid = 1/ (1+torch.exp(-x))
    return x*sigmoid

swish_layer = Swish()
x = torch.linspace(-5,5,200)
y=swish_layer(x)
```

```python
y
```

```python
plt.plot(x.numpy(),y.numpy())
plt.xlabel('X')
plt.ylabel('Swish of X')
plt.title('Swish Activation Function')
plt.grid(True)
plt.show()
```

Gaussian Error Linear Unit (GELU)

```python
import math
class GELU(nn.Module):
  def __init__(self):
    super(GELU, self).__init__()

  def forward(self, x):
    return 0.5*x*(1+torch.tanh(math.sqrt(2/math.pi)*(x+0.044715*x**3)))


gelu_layer = GELU()
x = torch.linspace(-5,5,200)
y=gelu_layer(x)
```

```python
y
```

```python
plt.plot(x.numpy(),y.numpy())
plt.xlabel('X')
plt.ylabel('GELU of X')
plt.title('GELU Activation Function')
plt.grid(True)
plt.show()
```
