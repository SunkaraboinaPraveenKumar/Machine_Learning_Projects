```python
import threading
from transformers import AutoTokenizer, AutoModelForCausalLM, TextIteratorStreamer
from time import time

tokenizer = AutoTokenizer.from_pretrained('gpt2')
model = AutoModelForCausalLM.from_pretrained('gpt2')

prompt = "The next day is bright"
tokens = tokenizer.encode(prompt, return_tensors='pt')

def stream_output(use_cache):
  streamer = TextIteratorStreamer(tokenizer, skip_special_tokens=True)

  thread = threading.Thread(
      target=model.generate,
      kwargs={
          'input_ids':tokens,
          "max_new_tokens":100,
          "use_cache":use_cache,
          "streamer":streamer,
      }
  )

  thread.start()

  start_time = time()
  for token in streamer:
    print(token, end="",flush=True)

  end_time = time()
  thread.join()

  elapsed = end_time-start_time
  print(f"\n\nUse Cache = {use_cache}, Time taken: {elapsed:.3f} seconds\n")

print("===With KV Caching===")
stream_output(use_cache=True)

print("===Without KV Caching===")
stream_output(use_cache=False)
```

```python
import threading
from transformers import AutoTokenizer, AutoModelForCausalLM, TextIteratorStreamer
from time import time

tokenizer = AutoTokenizer.from_pretrained('gpt2')
model = AutoModelForCausalLM.from_pretrained('gpt2')

prompt = "The next day is bright"
tokens = tokenizer.encode(prompt, return_tensors='pt')

def stream_output(use_cache):
  streamer = TextIteratorStreamer(tokenizer, skip_special_tokens=True)
  generated_text = ""  # Store the generated text

  thread = threading.Thread(
      target=model.generate,
      kwargs={
          'input_ids':tokens,
          "max_new_tokens":100,
          "use_cache":use_cache,
          "streamer":streamer,
      }
  )

  thread.start()

  start_time = time()
  for token in streamer:
    generated_text += token  # Append to generated_text

  end_time = time()
  thread.join()

  elapsed = end_time-start_time
  print(f"Use Cache = {use_cache}, Time taken: {elapsed:.3f} seconds\n")
  return generated_text  # Return the generated text

print("===With KV Caching===")
stream_output(use_cache=True)

print("===Without KV Caching===")
stream_output(use_cache=False)
```
